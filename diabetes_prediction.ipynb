```python
class DatasetAnalyzer:
    def __init__(self, dataset_path):
        self.dataset_path = dataset_path
        self.data = pd.read_csv(dataset_path)  # Assuming the dataset is in CSV format
    
    def info(self):
        return self.data.info()
    
    def describe(self):
        return self.data.describe()
    
    def head(self, n=5):
        return self.data.head(n)
    
    # You can add more methods for other operations you want to perform on the dataset


```

```python
dataset=DatasetAnalyzer('data/diabetes_binary_health_indicators_BRFSS2015.csv')
```
